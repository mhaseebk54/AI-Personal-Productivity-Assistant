```python
from gpt4all import GPT4All
model = GPT4All('mistral-7b-instruct-v0.1.Q4_0.gguf')
```

```python
name = input("Enter your name: ")
goal1 = input("Goal 1: ")
goal2 = input("Goal 2: ")
goal3 = input("Goal 3: ")
study_hours = input("How many study hours do you have today? ")
focus_topic = input("What subject or topic do you want to focus on? ")

goals = f"{goal1}, {goal2}, {goal3}"
```

```python
prompt_plan = f"""
Create a personalized productivity plan for a student named {name}.
Goals: {goals}.
Available study hours: {study_hours}.
Focus topic: {focus_topic}.

Include:
- Time blocks
- Study plan
- Break suggestions
- Task priority (High/Medium/Low)
- Tips for staying consistent
"""

print("\nGenerating your personalized plan...\n")

plan = model.generate(prompt_plan, max_tokens=300)
print("\n Personalized Day Plan \n")
print(plan)
```

```python
prompt_summary = f"Summarize the following plan :\n\n{plan}"

summary = model.generate(prompt_summary, max_tokens=200)

print("\n Summary  \n")
print(summary)
```

```python
prompt_motivation = "Give 3 short motivational lines for a student starting their daily routine."

motivation = model.generate(prompt_motivation, max_tokens=60)

print("\n Motivation for You \n")
print(motivation)
```

```python
ask_tip = input("\nDo you want a productivity tip? (yes/no): ").lower()

if ask_tip == "yes":
    tip_prompt = "Give 1 actionable productivity tip for a student."
    tip = model.generate(tip_prompt, max_tokens=80)
    print("\n Productivity Tip \n")
    print(tip)
```
